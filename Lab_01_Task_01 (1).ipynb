```python
import matplotlib.pyplot as plt
import numpy as np


data_exponential = np.random.exponential(scale=1.0, size=100000)
data_uniform = np.random.rand(100000)
data_normal = np.random.randn(100000)

plt.hist(data_exponential, density=True, bins=100, histtype="step", color="red", label="exponential")
plt.hist(data_uniform, density=True, bins=100, histtype="step", color="blue", label="uniform")
plt.hist(data_normal, density=True, bins=100, histtype="step", color="green", label="normal")
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Upload the file
from google.colab import files
uploaded = files.upload()

# Load the dataset with a different encoding if needed
try:
    data = pd.read_csv('Pokemon.csv', encoding='utf-8')
except UnicodeDecodeError:
    data = pd.read_csv('Pokemon.csv', encoding='latin1')

# Select the relevant features
features = data[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]

# Calculate Pearson and Spearman correlations
pearson_corr = features.corr(method='pearson')
spearman_corr = features.corr(method='spearman')

# Visualize the correlations using heatmaps
fig, ax = plt.subplots(1, 2, figsize=(14, 6))
sns.heatmap(pearson_corr, annot=True, ax=ax[0], cmap='coolwarm', fmt='.2f')
ax[0].set_title('Pearson Correlation')
sns.heatmap(spearman_corr, annot=True, ax=ax[1], cmap='coolwarm', fmt='.2f')
ax[1].set_title('Spearman Correlation')
plt.show()
```
